```python



import pandas as pd

# Sample Mall Customers Dataset
data = {
    'CustomerID': range(1, 201),
    'Gender': ['Male','Female']*100,
    'Age': [i%50 + 18 for i in range(200)],
    'Annual Income (k$)': [i%70 + 15 for i in range(200)],
    'Spending Score (1-100)': [i%100 + 1 for i in range(200)]
}

df = pd.DataFrame(data)

# Save to CSV
df.to_csv("Mall_Customers.csv", index=False)

print("Mall_Customers.csv file created successfully!")
```

```python
# Step 2: Load dataset
# (If you don’t have mall_customers.csv, download from Kaggle)
df = pd.read_csv("Mall_Customers.csv")

# Step 3: Select features (Annual Income & Spending Score)
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Step 4: Apply KMeans
kmeans = KMeans(n_clusters=5, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Step 5: Plot the clusters
plt.figure(figsize=(8,6))
plt.scatter(X.iloc[:,0], X.iloc[:,1], c=y_kmeans, cmap='rainbow', s=50)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            color='black', marker='X', s=200, label='Centroids')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segments')
plt.legend()
plt.show()

# Step 6: Print cluster count and inertia
print("Number of clusters:", kmeans.n_clusters)
print("Inertia:", kmeans.inertia_)
```
